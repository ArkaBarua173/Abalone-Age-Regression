    <li style='font-size: 12pt;'><b>Aroma: </b>Refers to the scent or fragrance of the coffee.</li>

    <li style='font-size: 12pt;'><b>Aftertaste: </b>Refers to the lingering taste that remains in the mouth after swallowing the coffee.</li>
    <li style='font-size: 12pt;'><b>Acidity: </b>Refers to the brightness or liveliness of the taste.</li>
    <li style='font-size: 12pt;'><b>Body: </b>Refers to the thickness or viscosity of the coffee in the mouth</li>
    <li style='font-size: 12pt;'><b>Balance: </b>Refers to how well the different flavor components of the coffee work together.</li>
    <li style='font-size: 12pt;'><b>Uniformity: </b>Refers to the consistency of the coffee from cup to cup.</li>
    <li style='font-size: 12pt;'><b>Clean.Cup: </b>Refers to a coffee that is free of any off-flavors or defects, such as sourness, mustiness, or staleness.</li>
    <li style='font-size: 12pt;'><b>Sweetness: </b>Refers to the pallet of sweetness offered by its taste.</li>
    <li style='font-size: 12pt;'><b>Moisture: </b>Represents the moisture content of the coffee beans, typically measured as a percentage.</li>
    <li style='font-size: 12pt;'><b>Quakers: </b>Indicates the presence of quaker beans, which are unripe or defective beans that fail to roast properly.</li>

# 🛠️ Data Preprocessing

```python
from sklearn.impute import SimpleImputer

numeric_imputer = SimpleImputer(missing_values=np.nan, strategy='mean').set_output(transform='pandas')
features = numeric_imputer.fit_transform(features)
```

```python
features.isna().sum()
```

```python
features.head()
```

```python
from sklearn.decomposition import PCA

pca = PCA().fit(features)
```

```python
exp_var_cumul = np.cumsum(pca.explained_variance_ratio_)
exp_var_cumul
```

```python
fig = px.line(
    x=range(1, exp_var_cumul.shape[0] + 1),
    y=exp_var_cumul,
    labels={"x": "Number of Components", "y": "Cumulative Explained Variance"},
    title="Explained Variance by Components",
    markers=True
)

fig.update_traces(line=dict(dash='dash'))

fig.show()
```

***We will choose 3 components as they capture almost 93% of the total variance***

```python
pca = PCA(n_components=3).set_output(transform='pandas')
pca_3 = pca.fit_transform(features)
pca_3
```

<a id='kmeans'></a>
# 🔍 K-means Clustering

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 21):
    kmeans_pca = KMeans(n_clusters = i, init = 'k-means++', random_state=42)
    kmeans_pca.fit(pca_3)
    wcss.append(kmeans_pca.inertia_)
```

```python
fig = px.line(
    x=range(1, 21),
    y=wcss,
    labels={"x": "Number of Clusters", "y": "WCSS"},
    title="KMeans with PCA Clustering",
    markers=True,
    line_shape="linear"
)

fig.show()
```

- ***We can see that after cluster 6 inertia starts decreasing in a linear fashion.***
- ***We will choose 6 as optimal number of clusters.***

```python
kmeans_pca = KMeans(n_clusters = 6, init = 'k-means++', random_state=42)
kmeans_pca.fit(pca_3)
```

```python
selected_df = df.select_dtypes(include='object')
segmented_kmeans_pca = pd.concat([selected_df.reset_index(drop=True), pca_3], axis=1)
new_column_names = ["Component_" + str(i+1) for i in range(3)]
segmented_kmeans_pca.columns.values[-3:] = new_column_names
segmented_kmeans_pca['K-means_labels'] = kmeans_pca.labels_
segmented_kmeans_pca.head()
```

```python
palette = sns.color_palette("husl", 6)

plt.figure(figsize=(10, 8))
sns.scatterplot(segmented_kmeans_pca, x="Component_1", y="Component_2", hue='K-means_labels', palette = palette)
plt.title('Clusters by PCA components')
plt.show()
```

```python
palette = px.colors.qualitative.Plotly

# Plot the scatter plot with the specified palette
fig = px.scatter_3d(segmented_kmeans_pca, x="Component_1", y="Component_2", z="Component_3",
                    color='K-means_labels', color_discrete_sequence=palette[:6])

fig.update_layout(title='Clusters by PCA components')
fig.show()
```
